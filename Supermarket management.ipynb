```python
import tkinter as tk
import MySQLdb as sql
from datetime import datetime

billlist=[["Item","Quantity","Price","Total Price"]]
def billing():
    i=i_var.get()
    q=q_var.get()
    p=p_var.get()
    if q<=0  or p<=0 or i=="":
        t6=tk.Label(top,text="Please enter Valid inputs!")
        t6.place(x=50,y=350)
        b3['state']=tk.DISABLED
        clear()
    l=[]
    stock=['toothpaste','brush','chocolate','rice','soap','salt','sugar','dal','biscuits','apple','banana','orange','carrot','potato','tomato']
    if i.lower() in stock:
        tp=q*p
        l=[i,q,p,tp]
        billlist.append(l)
        clear_ip()
    else:
        global t4
        t4=tk.Label(top,text="Item out of stock!")
        t4.place(x=50,y=350)
        b3['state']=tk.DISABLED
        clear_ip()
    
    
def clear_ip():
    i_var.set("")
    q_var.set("0")
    p_var.set("0")
    
def clear():
    clear_ip()
    billlist=[]
    n_var.set("")
    tf1.config(text="")
    tf2.config(text="")
    tf3.config(text="")
    t4.config(text="")
    t5.config(text="")
    t6.config(text="")
    t7.config(text="")
    b3['state']=tk.NORMAL
    

def genbill():
    n=n_var.get()
    if n=="":
        t7=tk.Label(top,text="name Cannot be Empty!")
        t7.place(x=50,y=400)
        clear()
        b3['state']=tk.DISABLED
    ttp=0
    j=450
    iter=0
    global t5
    t5=tk.Label(top,text="Your Bill is ready!")
    t5.place(x=50,y=350)
    global tf1
    global tf2
    tf1=tk.Label(top,text="Item         Quantity   Price   Total price",font=("Arial", 10),fg='red')
    tf1.place(x=50,y=430)
    for i,q,p,tp in billlist:
        if iter!=0:
            tf2=tk.Label(top,text=str(i)+"              "+str(q)+"           "+str(p)+"            "+str(tp),font=("Arial", 10),fg='red')
            tf2.place(x=50,y=j)
            tp=int(tp)
            ttp+=tp
            tp=str(tp)
            j+=20
        iter+=1
    global tf3  
    tf3=tk.Label(top,text="Total Bill: "+str(ttp)+" INR",font=("Arial", 20),fg='red')
    tf3.place(x=50,y=650)
    d=datetime.now()
    d=d.strftime("%d/%m/%Y %H:%M:%S")
    con=sql.connect(host="localhost",user="root",password="",database="salesdb")
    c=con.cursor()
    q="insert into sales(date_time,name,total_bill) values(%s,%s,%s)"
    val=(d,n,str(ttp))
    c.execute(q,val)
    con.commit()
    con.close()

top=tk.Tk()
top.title("Billing")
top.geometry("600x700")
top.config(bg='black')

i_var=tk.StringVar()
q_var=tk.IntVar()
p_var=tk.IntVar()
n_var=tk.StringVar()
q_var.set("0")
p_var.set("0")
i_var.set("")

t=tk.Label(top,text="Enter the item, quantity and price per peice", bg='yellow',font=("Helvetica", 20),fg='purple')
t.place(x=40,y=50)

t0=tk.Label(top,text="Customer name: ",font=("Arial Bold", 15),bg='black',fg='lavender')
t0.place(x=50,y=100)

t1=tk.Label(top,text="Item: ",font=("Arial Bold",15),bg='black',fg='lavender')
t1.place(x=50,y=150)

t2=tk.Label(top,text="Quantity(pc/kg): ",font=("Arial Bold", 15),bg='black',fg='lavender')
t2.place(x=50,y=200)

t3=tk.Label(top,text="Price: ",font=("Arial Bold", 15),bg='black',fg='lavender')
t3.place(x=50,y=250)

ip0=tk.Entry(top,textvariable=n_var,font=("Arial", 15),bg='sky blue')
ip0.place(x=220,y=100)

ip1=tk.Entry(top,textvariable=i_var,font=("Arial", 15),bg='sky blue')
ip1.place(x=220,y=150)

ip2=tk.Entry(top,textvariable=q_var,font=("Arial", 15),bg='sky blue')
ip2.place(x=220,y=200)

ip3=tk.Entry(top,textvariable=p_var,font=("Arial", 15),bg='sky blue')
ip3.place(x=220,y=250)

b1=tk.Button(top,text="Save Item",bg='yellow',fg='black',command=billing)
b1.place(x=100,y=300)

b2=tk.Button(top,text="Clear",bg='red',fg='white',command=clear)
b2.place(x=200,y=300)

b3=tk.Button(top,text="Generate Bill",bg='dark green',fg='white',command=genbill)
b3.place(x=300,y=300)
top.mainloop()
```
